**PROBLEMA 4 ITERACIÓN DE PUNTO FIJO**

```python
from math import *
import numpy as np
import matplotlib . pyplot as plt
```

```python
def funcion4(x):
    return x**3-x-1

def g4(x):
  return pow(x+1,1/3)
```

```python
# Ingreso datos de entrada para los diferentes métodos a trabajar
a = 1
b = 2

# guarda valores iniciales
a0 = a
b0 = b

tol = 0.0001 
nmax = 100 
error = 100
niter = 0
```

```python
#iteracion de punto fijo
m=1

#Evaluacion de la función en los puntos a, b y m
fa = funcion4(m)
fm = g4(m)

print("# iter\t\t p0 \t\t g(p0) \t\t f(p0) \t\t error")
print("{0} \t\t {1:6.4f} \t {2:6.4f} \t {3:6.4f} \t {4:6.4f}".format(niter, m, fm, fa, error))

# ciclo iterativo
m=g4(a0)
while niter == nmax:
    a=m
    fm= g4(a)
    fa= funcion4(m)
    error = abs(fa-fm)
    niter += 1
    print("{0} \t\t {1:6.6f} \t {2:6.6f} \t {3:6.6f} \t {4:6.6f}".format(niter, m, fm, fa, error ))

print("La raíz de la función dada en el intervalo [{0:6.4f},{1:6.4f}] es {2:6.7f}".format(a0,b0,m))
```

```python
def f1(x):
    return np.cos(x)-x*np.exp(x)
def f2(x):
    return x
def f3(x):
    return np.cos(x)/np.exp(x)

x=np.linspace(0, 1, 70)
```

```python
plt.plot(x, [f1(i) for i in x],color="red", label=r'y= $\cos(x)-xe^{x}$')
plt.plot(x, [f2(i) for i in x],color="blue", label='y=x')
plt.plot(x, [f3(i) for i in x],color="green", label=r'y= $\frac{cos(x)}{e^{x}}$')
plt.legend(loc = 'upper right')
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.axvline(0.517757, color="grey", linestyle='dotted')
plt.axhline(0.517757, color="grey", linestyle='dotted')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title(r'f(x)= $\cos(x)-xe^{x}$')
plt.grid()
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```

```python
from math import *
import numpy as np
import matplotlib . pyplot as plt
def funcion1(x):
    return x**3 + x -4
# Ingreso datos de entrada para los diferentes métodos a trabajar
a = 1
b = 2

# guarda valores iniciales
a0 = a
b0 = b


tol = 0.0001 
nmax = 100 
error = 100
niter = 0


```

```python
# Método de Bisección
# evaluo primer valor medio
m = a + (b - a)/2

#Evaluacion de la función en los puntos a, b y m
fa = funcion1(a)
fb = funcion1(b)
fm = funcion1(m)

print("# iter\t\t a \t\t f(a) \t\t b \t\t f(b) \t\t m \t\t f(m) \t\t error")
print("{0} \t\t {1:6.4f} \t {2:6.4f} \t {3:6.4f} \t {4:6.4f} \t {5:6.4f} \t {6:6.4f} \t {7:6.4f}".format(niter, a0, fa, b0, fb, m, fm, error ))

# ciclo iterativo
while error > tol and niter < nmax:
    m = a + (b - a) / 2
    if np.sign(fa) == np.sign(fm):
        a = m
        fa = funcion1(a)
    else:
        b = m
        fb = funcion1(b)
    
    m = a + (b - a)/2
    fm = funcion1(m)
    error = abs(b - a)
    niter += 1
    print("{0} \t\t {1:6.6f} \t {2:6.6f} \t {3:6.6f} \t {4:6.6f} \t {5:6.6f} \t {6:6.6f} \t {7:6.6f}".format(niter, a, fa, b, fb, m, fm, error ))

print("La raíz de la función dada en el intervalo [{0:6.4f},{1:6.4f}] es {2:6.7f}".format(a0,b0,m))
```
